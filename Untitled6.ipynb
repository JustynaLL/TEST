```python
activities = ['Client Entertainment', 'Client Sponsored Event','Event - Attended','Relationship Meeting','Training and Seminar - Delivered']
```

```python
df2 = df[df["Activity Type"].isin(activities)].reset_index(drop=True)
```

```python
df2['Concat'] = df2['Activity Date']+df2['Summary']
```

```python
df2.head(3)
```

```python
df3 = df2.assign(Concat = lambda x: df2['Activity Date']+df2['Summary'])
```

```python
df3['Linklaters Contacts'] = df3['Linklaters Contacts'].fillna('Unknown')
df3['Client Contacts'] = df3['Client Contacts'].fillna('Unknown')
```

```python
df3_vlookup = df3.copy()
df3_vlookup['Linklaters Contacts'] = df3_vlookup['Linklaters Contacts'].str.split('; ')
```

```python
PG_dict = {}
PG = []
PG_main = []

for x in df3_vlookup['Linklaters Contacts']:
  for n in range(len(x)):
    x[n] = " ".join(x[n].split(", ")[::-1])
    for index in range(len(staff.Name)):
      if x[n] == staff.Name[index]:
        PG_dict.update({x[n] : staff['Practice/BS Group'][index]})
```

```python
for x in df3_vlookup['Linklaters Contacts']:
  PG=[]
  for n in range(len(x)):
    PG.append(PG_dict.get(x[n]))
  PG_main.append(set(PG))
```

```python
df3_vlookup['PG'] = PG_main
```

```python
df3_vlookup
```

```python
df4 = pd.DataFrame(df3.groupby('Concat',as_index=False).agg(lambda x: '; '.join(set(x))))
```

```python
df4['Client Contacts'] = df4['Client Contacts'].str.split('; ')
```

```python
for x in df4['Client Contacts']:
  for n in range(len(x)):
    x[n] = " ".join(x[n].split(", ")[::-1])

for n in range(len(df4['Client Contacts'])):
  if len(df4['Client Contacts'][n]) in range(2,5):
    df4['Client Contacts'][n] = ', '.join(df4['Client Contacts'][n])
    k = df4['Client Contacts'][n].rfind(',')
    df4['Client Contacts'][n] = df4['Client Contacts'][n][:k] + ' and' + df4['Client Contacts'][n][k+1:]
  elif len(df4['Client Contacts'][n]) > 4:
    df4['Client Contacts'][n] = 'a few members from your team'
  else:
    df4['Client Contacts'][n] = ', '.join(df4['Client Contacts'][n])
```

```python
df4['Linklaters Contacts'] = df4['Linklaters Contacts'].str.split('; ')

for x in df4['Linklaters Contacts']:
  for n in range(len(x)):
    x[n] = " ".join(x[n].split(", ")[::-1])

for n in range(len(df4['Linklaters Contacts'])):
  if len(df4['Linklaters Contacts'][n]) in range(2,5):
    df4['Linklaters Contacts'][n] = ', '.join(df4['Linklaters Contacts'][n])
    k = df4['Linklaters Contacts'][n].rfind(',')
    df4['Linklaters Contacts'][n] = df4['Linklaters Contacts'][n][:k] + ' and' + df4['Linklaters Contacts'][n][k+1:]
  elif len(df4['Linklaters Contacts'][n]) > 4:
    df4['Linklaters Contacts'][n] = 'a few of our lawyers'
  else:
    df4['Linklaters Contacts'][n] = ', '.join(df4['Linklaters Contacts'][n])
```

```python
df4 = df4.merge(staff, left_on='Linklaters Contacts', right_on='Name',how='left')
df4['Individual'] = df4['Individual'].astype('Int32',errors='ignore')
```

```python
df4
```

```python
from datetime import datetime
```

```python
def final():
  narratives = {}

  for x in range(len(df4)):
    if df4['Activity Type'][x] in(['Client Entertainment', 'Relationship Meeting']):  
      narratives.update({df4['Concat'][x] : f'{df4["Linklaters Contacts"][x][0].upper() + df4["Linklaters Contacts"][x][1:]} met with {df4["Client Contacts"][x]} on {datetime.strptime(df4["Activity Date"][x], "%d-%m-%Y").date().strftime("%-d %B")}'})
    elif df4['Activity Type'][x] in (['Training and Seminar - Delivered','Client Sponsored Event']):
      narratives.update({df4['Concat'][x] : f'{df4["Linklaters Contacts"][x][0].upper() + df4["Linklaters Contacts"][x][1:]} attended the {df4["Summary"][x]} on {datetime.strptime(df4["Activity Date"][x], "%d-%m-%Y").date().strftime("%-d %B")}'})
    else:
      narratives.update({df4['Concat'][x] : f'{df4["Client Contacts"][x][0].upper() + df4["Client Contacts"][x][1:]} attended the {df4["Summary"][x]} on {datetime.strptime(df4["Activity Date"][x], "%d-%m-%Y").date().strftime("%-d %B")}'})
  return narratives
```

```python
final()
```

```python
df_narratives = pd.DataFrame.from_dict(final(),orient='index',columns=['Narrative'])
```

```python
df_final = df4.merge(df_narratives, left_on='Concat', right_index = True,how='left')
```
